#### Devolver qué personajes están en todos los episodios

```python
import json

def repeated_characters():
    enderezo = "https://rickandmortyapi.com/api/episode/"
    resposta = requests.get(enderezo)
    variables=json.loads(json.dumps(resposta.json()))

    ret = []
    nombres = variables["results"][0]["characters"]

    while True:
        for result in variables["results"]:
            for n in nombres:
                if n not in result["characters"]:
                    nombres.remove(n)

        if variables["info"]["next"] != None:
            enderezo = variables["info"]["next"]
            resposta = requests.get(enderezo)
            variables=json.loads(json.dumps(resposta.json()))
        else:
            for n in nombres:
                resposta = requests.get(n)
                variables=json.loads(json.dumps(resposta.json()))
                ret.append(variables["name"])

            return ret

print(repeated_characters())
```

#### Devolver el nombre de todos los episodios y de todos los personajes

```python
import json

def recoger_datos(enderezo):
    resposta = requests.get(enderezo)
    variables=json.loads(json.dumps(resposta.json()))
    name = []

    while True:
        for result in variables["results"]:
            if result["name"]!=None:
                name.append(result["name"])

        if variables["info"]["next"] != None:
            enderezo = variables["info"]["next"]
            resposta = requests.get(enderezo)
            variables=json.loads(json.dumps(resposta.json()))
        else: return name

a = recoger_datos("https://rickandmortyapi.com/api/episode/")
b = recoger_datos("https://rickandmortyapi.com/api/character")

print("Episodios: ",a)
print("Personajes:",b)
```
