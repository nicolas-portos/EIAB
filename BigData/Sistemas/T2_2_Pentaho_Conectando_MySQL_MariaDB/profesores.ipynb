# Ampliación 1 - Sacando datos de una tabla HTML y limpiarlos

```python
import pandas as pd
import html5lib
```

### Descargar la tabla de profesores substitutos

```python
df = pd.read_html("https://www.edu.xunta.gal/substitutoslistas/SubstitucionsAdxudicadas.do?corpo=590&especialidade=227&dataFin=&nif=&idDoc=1&sort=dataAlta&provincia=&dir=asc&DIALOG-EVENT-substitucionsAdxudicadas=Buscar&searchid=&centro=&dataIni=")[0]
df
```

- Creamos una nueva table llamada "Cod_Centro" donde insertaremos los codigos de centro localizados en la columna "Centro"

```python
datos = df["Centro"].str.split("-",n=1, expand=True)
df.insert(3, "Cod_Centro", datos[0])
df
```

- Eliminamos los valores de cod_centro de la columna centro, ya que al haber hecho un split cogemos solo el nombre del centro

```python
df["Centro"] = datos[1]
df
```

- Eliminamos los espacios y las barras por "_"

```python
df.columns = df.columns.str.replace(" ","_")
df.columns = df.columns.str.replace("/","_")
df
```

- Eliminamos los acentos

```python
df.columns = df.columns.str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8')
df
```

- Exportamos a csv

```python
df.to_csv("archives/profesores.csv")
```
