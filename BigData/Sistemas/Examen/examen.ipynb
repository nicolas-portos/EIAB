## 2- Hacer dataframe en base a la api de rick y morty

```python
import json
import pandas as pd

def recoger_locations_api():
    enderezo = "https://rickandmortyapi.com/api/location"
    resposta = requests.get(enderezo)
    variables = json.loads(json.dumps(resposta.json()))
    ret = []

    while True:
        for result in variables["results"]:
            char = [1,2]
            try:
                resident = result["residents"]
                r = requests.get(resident[0])
                v = json.loads(json.dumps(r.json()))
                char[0] = v["name"]
                char[1] = v["species"]
            except:
                char[0] = "empty"
                char[1] = "empty"

            ret.append({"id": result["id"],
                   "name": result["name"],
                   "type": result["type"],
                   "dimension": result["dimension"],
                   "resident1-name": char[0],
                   "resident1-species": char[1]
                   })

        if variables["info"]["next"] != None:
             enderezo = variables["info"]["next"]
             resposta = requests.get(enderezo)
             variables=json.loads(json.dumps(resposta.json()))
        else: 
            return ret

datos = recoger_locations_api()
```

```python
df = pd.DataFrame(datos)
df
```

```python
df.set_index(df["id"])
df
```

## 3- Exportar dataframe a JSON, CSV, PARQUE

```python
df.to_json("archives/rick_and_morty_locations.json")
df.to_csv("archives/rick_and_morty_locations.csv")
df.to_parquet("archives/rick_and_morty_locations.parquet")
```
