Limpieza

```python
df.drop(["_id"],axis=1,inplace=True)
df["year"] = df["year"].astype("str").str[:4]
```

```python
def contar_valores(df:pd.DataFrame, column:str):
    df2 = df[df[column].notna()]
    ret = pd.Series([item for sublist in df2[column] for item in sublist]).value_counts()
    # return pd.Series([item for sublist in df[column] if isinstance(sublist, list) for item in sublist]).value_counts()
    return ret
```

D) Mirar se hai correlación entre imdb e rotten tomatoes.

```python
limdb = []
ltomatoes = []

df_r = df.dropna()

for imdb, tomatoes in zip(df_r["imdb"],df_r["tomatoes"]):
    if "viewer" in tomatoes:
        if "rating" in tomatoes["viewer"]:
            ltomatoes.append(tomatoes["viewer"]["rating"])
            limdb.append(imdb["rating"])

dfCorr = pd.DataFrame({"tomatoes":ltomatoes,"imdb":limdb})
dfCorr.replace(r'^\s*$', None, regex=True, inplace=True)
dfCorr["imdb"].astype("float")
dfCorr.corr()
```
