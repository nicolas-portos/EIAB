# Ejercicio

```python
import pandas as pd
import os

folder = "archives/"
df = pd.DataFrame()
for file in os.listdir(folder):
    path = folder+file
    df = pd.concat([df,pd.read_excel(path,usecols=lambda x: 'Unnamed' not in x, skiprows=7)])

df
```

## Limpieza de datos

### Limpiamos los nombres que sean nulos

```python
df.dropna(subset=["NOMBRE"],inplace=True)
df.info()
```

### Normalizamos los nombres para posteriormente utilizarlos de forma más sencilla

```python
# datos = df["NOMBRE"].str.upper().str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8').str.replace(" ","").str.replace(",","")
datos = df["NOMBRE"].str.upper().str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8').str.replace(",","").str.lower()
df.insert(4, "Nombre_Normalizado", datos)
df
```

## Busquedas Relacionadas con el mandato

### Mostrar alcaldes que hayan repetido mandato

```python
df["Nombre_Normalizado"].value_counts()[df["Nombre_Normalizado"].value_counts()>1].to_frame(name="Total_Legislaturas")
```

### Bajo que partidos

```python
df.groupby(["Nombre_Normalizado", "LISTA"]).size().to_frame(name="Legislaturas_por_partido").sort_values(by="Legislaturas_por_partido",ascending=False)
```

### Alcaldes que siguen al mando

```python
df[df["FECHA BAJA"].isna()]
```

### Que alcaldes no siguen en el mando

```python
df[df["FECHA BAJA"].notna()]
```

## Datos interesantes

### Número de alcaldes (por partido) por provincia

```python
df.groupby(["PROVINCIA", "LISTA"]).size().to_frame("N. Alcaldes")
```

### Mostrar el partido que más alcaldes ha tenido por municipio y el número de alcaldes

```python
df_m = df.groupby(["MUNICIPIO", "LISTA"]).size().to_frame("Max Alcaldes Municipio").sort_values(by="MUNICIPIO",ascending=False).reset_index()

df_m.sort_values(["LISTA","Max Alcaldes Municipio"],ascending=[True,False]).drop_duplicates(["MUNICIPIO"]).sort_values(["Max Alcaldes Municipio"],ascending=False)
```
