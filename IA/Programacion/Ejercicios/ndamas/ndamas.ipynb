```python
import numpy as np
import random
from IPython.display import clear_output

"""
Generar una población inicial
durante un número máximo de iteraciones:
  ordenar la poblacion
  quedarse con los mejores
  si alguno es solucion acabar
  cruzar individuos
  añadir hijos a la población
"""
```

```python
class ndamas():
    
    poblacion = []
    n_damas = 0

    def __init__(self, n:int):
        if n < 3: raise Exception("No existe solución para ndamas si el tablero es 2x2 o menor")
        self.n_damas = n

    def mostrar_tableros(self):
        # for tablero in self.poblacion:
        #     self.mostrar_tablero(tablero)
        [self.mostrar_tablero(tablero) for tablero in self.poblacion]

    def mostrar_tablero(self,tablero):
        # for fila in tablero:
        #     print(fila,end="\n")
        [print(fila,end="\n") for fila in tablero]
        print()

    # valora cuantos errores tiene la posición
    # 0 es ningún error, cualquier otro valor es positivo
    def valora_errores_posicion(self, tab):         
        nerrores = 0
        tablero = np.array(tab)
        tablero_inverso = np.array(tablero).T

        for i in range(len(tablero)):
            if sum(tablero[i]) > 1: 
                nerrores+= sum(tablero[i])

            if sum(tablero_inverso[i]) > 1: 
                nerrores += sum(tablero_inverso[i])

        diags = [tablero[::-1,:].diagonal(i) for i in range(-tablero.shape[0]+1, tablero.shape[1])]
        diags.extend(tablero.diagonal(i) for i in range(tablero.shape[1]-1, -tablero.shape[0],-1))

        nerrores += sum([sum(n.tolist()) if sum(n.tolist()) > 1 else 0 for n in diags])
        # for n in diags: 
            # if sum(n.tolist()) > 1: nerrores += sum(n.tolist())
        
        return nerrores
    
    # generar n_damas individuos al azar
    def genera_nelementos(self, elementos:int = 1):

        def generar_tableros(n:int):
            self.poblacion = []
            for _ in range(n):
                # Dado que este problema no puede tener más damas que columnas y lineas,
                # hacemos que el numero de damas sea igual al numero de lineas y columas
                # si tenemos 8 damas el tablero sera de 8x8
                self.poblacion.append(np.zeros((self.n_damas,self.n_damas),dtype=list).tolist())

        def asignar_damas_tablero():
            for tablero in self.poblacion:
                while sum(list(map(sum, tablero))) != self.n_damas:
                    tablero[random.randint(0,len(tablero)-1)][random.randint(0,len(tablero)-1)] = 1

        def errores_poblacion():
            errores = []
            for tablero in self.poblacion:
                errores.append(self.valora_errores_posicion(tablero))
            # Devuelve una lista con el numero de errores, el index de esta lista coincide 
            # con el de self.poblacion
            return errores
        
        def cruzar_genes():
            nueva_poblacion = []

            self.poblacion.sort(key=self.valora_errores_posicion)
            self.poblacion = self.poblacion[:(int(len(self.poblacion)//2))]

            # Crear nueva poblacion en base a la anterior aleatoriamente
            while len(nueva_poblacion) < len(self.poblacion):
                tab1 = self.poblacion[(random.randint(0, len(self.poblacion)-1))]
                tab2 = self.poblacion[(random.randint(0, len(self.poblacion)-1))]
                tab_hijo = []

                for i in range(len(tab1)):
                    fila1 = tab1[i]
                    fila2 = tab2[i]
                    if random.randint(0,1) == 0:
                        tab_hijo.append(fila1)
                    else:
                        tab_hijo.append(fila2)
                if(np.count_nonzero(np.array(tab_hijo)) == self.n_damas):
                    nueva_poblacion.append(tab_hijo)

            ##### Añadir 50% aleatorios
            n = int(len(self.poblacion) * 0.50)
            n_pob = []
            for _ in range(n):
                n_pob.append(np.zeros((self.n_damas,self.n_damas),dtype=list).tolist())
            
            for tablero in n_pob:
                while sum(list(map(sum, tablero))) != self.n_damas:
                    tablero[random.randint(0,len(tablero)-1)][random.randint(0,len(tablero)-1)] = 1

            self.poblacion += nueva_poblacion + n_pob
            
        generar_tableros(elementos)
        asignar_damas_tablero()
        errores = errores_poblacion()
        contador = 0
        
        while True:
            contador+=1

            if 0 in errores:
                clear_output()
                print("Has ganado, Te ha llevado:", contador, "ciclos")

                index = errores.index(0)
                tablero = self.poblacion[index]
                self.mostrar_tablero(tablero)
                break

            cruzar_genes()
            errores.sort()
            print("Error max",errores[0])
            errores = errores_poblacion()
```
