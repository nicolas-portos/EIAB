```python
TEST = "dump/"
IMAGES_FOLDER = "images"
lgestures = ["none","piedra","papel","tijera"]

# create gestures folders
for gesture in lgestures:
    if not os.path.exists(f"{IMAGES_FOLDER}"): os.mkdir(f"{IMAGES_FOLDER}")
    if not os.path.exists(f"{IMAGES_FOLDER}/{gesture}"):
        os.mkdir(f"{IMAGES_FOLDER}/{gesture}")

count = 0
for user in os.listdir(TEST):
    for gesture in os.listdir(f"{TEST}/{user}/"):
        images = os.listdir(f"{TEST}/{user}/{gesture}/")
        for image in images:
            count += 1
            shutil.copy(f"{TEST}/{user}/{gesture}/{image}", f"{IMAGES_FOLDER}/{gesture}/{gesture}_{count}.jpg")

            # Si existe esa imagen para no sobreescribirla añadele un 1
            # if os.path.exists(f"{IMAGES_FOLDER}/{gesture}/{image}"): 
            #     shutil.copy(f"{TEST}/{user}/{gesture}/{image}", f"{IMAGES_FOLDER}/{gesture}/1{image}")
            # else: shutil.copy(f"{TEST}/{user}/{gesture}/{image}", f"{IMAGES_FOLDER}/{gesture}/{image}")
```

```python
NUM_EXAMPLES = 4

for gesture in os.listdir("images/"):
    example_filenames = os.listdir(f"images/{gesture}/")[:NUM_EXAMPLES]
    fig, axs = plt.subplots(1, NUM_EXAMPLES, figsize=(10,2))
    for i in range(NUM_EXAMPLES):
        axs[i].imshow(plt.imread(os.path.join(f"images/{gesture}/", example_filenames[i])))
        axs[i].get_xaxis().set_visible(False)
        axs[i].get_yaxis().set_visible(False)
    fig.suptitle(f'Showing {NUM_EXAMPLES} examples for {gesture}')

plt.show()
```

```python
for gesture in os.listdir("images/"):
    for image in os.listdir(f"images/{gesture}"):
        img = cv.imread(f"images/{gesture}/{image}")

        flipped = cv.flip(img,0)
        cv.imwrite(f"images/{gesture}/f0{image}", flipped)
        flipped = cv.flip(img,1)
        cv.imwrite(f"images/{gesture}/f1{image}", flipped)
        th = cv.transpose(img, 0)
        cv.imwrite(f"images/{gesture}/t0{image}", th)
        th = cv.transpose(img, 1)
        cv.imwrite(f"images/{gesture}/t1{image}", th)
        cv.imwrite(f"images/{gesture}/{image}", img)
```

```python
data = gesture_recognizer.Dataset.from_folder(
    dirname="images/",
    hparams=gesture_recognizer.HandDataPreprocessingParams()
)
train_data, rest_data = data.split(0.8)
validation_data, test_data = rest_data.split(0.5)
```

```python
hparams = gesture_recognizer.HParams(export_dir="exported_model")
options = gesture_recognizer.GestureRecognizerOptions(hparams=hparams)
model = gesture_recognizer.GestureRecognizer.create(
    train_data=train_data,
    validation_data=validation_data,
    options=options
)
```

```python
loss, acc = model.evaluate(test_data, batch_size=1)
print(f"Test loss:{loss}, Test accuracy:{acc}")
```

```python
model.export_model()
```
