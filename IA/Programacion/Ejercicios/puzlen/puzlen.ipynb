```python
import matplotlib.pyplot as plt
import numpy as np
import math, json, pprint

from Grafo import Grafo

class PuzleN(Grafo):

    def __init__(self, np=8):
        Grafo.__init__(self)
        self.lado = int(math.sqrt(np+1))
    
    def es_solucion(self, nodo_actual,nodo_destino):
        return nodo_actual == nodo_destino

    def calcula_distanciaDst(self, nodo, nodo_destino):
        l_nodo = nodo.split("-")
        l_nodo_destino = nodo_destino.split("-")
        return len([i for i,j in zip(l_nodo, l_nodo_destino) if(i!=j)])
        # ret = 0
        # for i,j in zip(l_nodo, l_nodo_destino):
        #     if(i!=j): ret +=1
        # return ret
        # return np.count_nonzero(np.not_equal([l_nodo],[l_nodo_destino]))
        
    # devuelve una lista de nodos hijo
    def genera_sucesores(self, nodo):
      hijos = []

      l = nodo.split("-")
      ind = l.index("0")
      fila = ind // self.lado;  columna = ind % self.lado
      incs = [(-1,0), (1,0), (0,-1), (0,1)]

      for inc in incs:
         f = fila + inc[0]; c = columna + inc[1]
         if f >= 0 and f < self.lado and c >= 0 and c < self.lado:
            #intercambiar
            l[ind] = l[f*self.lado + c];   l[f*self.lado + c] = "0"
            s = "-".join(l)
            hijos.append(s)
            # dejar todo como estaba
            l[f*self.lado + c] = l[ind];   l[ind] = "0"
      return hijos
    
    def pop_abiertos(self, modo):
       if modo == "avaricioso":
            valorBorrar = min(self.abiertos, key=lambda x: self.get_node_attribute(x ,"distanciaDst"))
            self.abiertos.remove(valorBorrar)
            ret = valorBorrar
            return ret
       else: return super().pop_abiertos(modo)
```
