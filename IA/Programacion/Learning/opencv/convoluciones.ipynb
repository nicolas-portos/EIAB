```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = cv.imread("archives/test.png")
src = cv.cvtColor(src, cv.COLOR_BGR2RGB)

plt.imshow(src)
```

```python
filtro = np.ones((5,5), np.float32)/25
img2 = cv.filter2D(src=src, ddepth=-1, kernel=filtro )
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(src)
ax2.imshow(img2)
```

```python
#desenfoques:
img2 = cv.blur(src=src, ksize=(5,5))
img3 = cv.GaussianBlur(src=src, ksize=(5,5), sigmaX=0, sigmaY=0 )
img4 = cv.medianBlur(src, ksize=5)
img5 = cv.bilateralFilter(src=src, d=9, sigmaColor=75, sigmaSpace=75)
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1,5)
ax1.imshow(src)
ax2.imshow(img2)
ax3.imshow(img3)
ax4.imshow(img4)
ax5.imshow(img5)
```

```python
img1 = cv.imread("archives/test2.jpg")
gray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
plt.imshow(gray1, cmap="gray")
```

```python
imgT = cv.GaussianBlur(src=gray1, ksize=(5,5), sigmaX=0, sigmaY=0 )
laplaciano = cv.Laplacian(imgT, cv.CV_64F)
sobel1 = cv.Sobel(imgT, cv.CV_64F, 1, 0, ksize=5)
sobel2 = cv.Sobel(imgT, cv.CV_64F, 0, 1, ksize=5)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
#ax1.imshow(imgT, cmap="gray")
ax1.imshow(laplaciano, cmap="gray")
ax2.imshow(sobel1, cmap="gray")
ax3.imshow(sobel2, cmap="gray")
```

```python
imgT = cv.GaussianBlur(src=gray1, ksize=(5,5), sigmaX=50, sigmaY=50 )
laplaciano = cv.Laplacian(imgT, cv.CV_64F)
resultado = imgT.astype(float)
resultado = resultado - laplaciano
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(imgT, cmap="gray")
ax2.imshow(resultado, cmap="gray")
```
