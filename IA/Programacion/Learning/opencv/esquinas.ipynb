```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
src = "archives/ajedrez.jpg"
img = cv.imread(src)
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

img1 = np.float32(gris)
dest = cv.cornerHarris(img1, 2, 5, 0.07)
dest = cv.dilate(dest, None)

maximo = dest.max()
umbral = maximo * 0.05
img3 = img2.copy()
img3[dest > umbral] = [0,0,255]
plt.imshow(img3, cmap = "gray")
```

```python
img = cv.imread(src)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
esquinas = cv.goodFeaturesToTrack(gris, 100, 0.3, 10)
esquinas = np.intp(esquinas)
for esquina in esquinas:
  x,y = esquina[0]
  cv.circle(img, (x,y), 3, (255, 0,0), -1)

plt.imshow(img)
```
