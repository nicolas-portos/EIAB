```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as nu
import pylab
```

```python
pylab.rcParams['figure.figsize'] = (8.0, 6.0)

src = cv.imread("archives/test.png")
src = cv.cvtColor(src, cv.COLOR_BGR2RGB)
plt.imshow(src)
```

```python
fig, axes = plt.subplots(1, 3)
fig.suptitle("R G B")
axes[0].imshow(src[:,:,0], cmap="gray", vmin=0, vmax=255)
axes[1].imshow(src[:,:,1], cmap="gray", vmin=0, vmax=255)
axes[2].imshow(src[:,:,2], cmap="gray", vmin=0, vmax=255)
```

```python
fig, axes = plt.subplots(1, 3)
fig.suptitle("Histograma R G B")
axes[0].plot(cv.calcHist([src], [0], None, [256], [0,255]), color = "red")
axes[1].plot(cv.calcHist([src], [1], None, [256], [0,255]), color = "green")
axes[2].plot(cv.calcHist([src], [2], None, [256], [0,255]), color = "blue")
```

```python
fig, axes = plt.subplots(1, 3)
fig.suptitle("R G B Equalizados")
axes[0].imshow(cv.equalizeHist(src[:,:,0]), cmap="gray", vmin=0, vmax=255)
axes[1].imshow(cv.equalizeHist(src[:,:,1]), cmap="gray", vmin=0, vmax=255)
axes[2].imshow(cv.equalizeHist(src[:,:,2]), cmap="gray", vmin=0, vmax=255)
```

```python
img_r = cv.equalizeHist(src[:,:,0])
img_g = cv.equalizeHist(src[:,:,1])
img_b = cv.equalizeHist(src[:,:,2])
merged = cv.merge([img_r, img_g, img_b])

fig, axes = plt.subplots(1 ,3)
fig.suptitle("Histograma R G B equalizado")
axes[0].plot(cv.calcHist([merged], [0], None, [256], [0,255]), color = "red")
axes[1].plot(cv.calcHist([merged], [1], None, [256], [0,255]), color = "green")
axes[2].plot(cv.calcHist([merged], [2], None, [256], [0,255]), color = "blue")
```
