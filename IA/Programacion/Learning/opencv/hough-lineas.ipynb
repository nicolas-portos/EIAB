```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import pandas as pd
```

```python
src = cv.imread("archives/carretera.jpg")
img = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
img2 = cv.cvtColor(src, cv.COLOR_BGR2RGB)
plt.imshow(img, cmap="gray")
```

```python
imgBlur = cv.GaussianBlur(img,(7,7),0)
edges = cv.Canny(imgBlur, 100, 70)
minVotos = 80
minLen = 200
maxGap = 20
lineas = cv.HoughLinesP(edges,1,np.pi/100,minVotos,minLen,maxGap)
for linea in lineas:
    x1,y1,x2,y2 = linea[0]
    cv.line(img2, (x1,y1), (x2,y2), (255,0,0),2)
plt.imshow(img2)
```
