Datos:
‚Ä¢ ùë• ‚àà [0, 2ùúã] (1000 puntos)
‚Ä¢ ùë¶ = ùëí‚àíùõºùë•ùë†ùëñùëõ(ùõΩùë•), donde ùõº = 0.7 y ùõΩ = 10.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 1000)
alfa = 0.7
beta = 10.0
y = np.exp(-alfa*x)*np.sin(beta*x)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
ax.plot(x, y, label = "exp(-alfa*x)*sin(beta*x)", linewidth=3, color="green")
ax.legend()
ax.grid()
```
